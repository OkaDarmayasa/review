```python
import csv
import json

#load json
with open('dict.json', 'r', encoding='utf-8') as file:
    thesaurus_data = json.load(file)

#buat list untuk menampung
words_with_only_antonim = []

#iterate file jsonnya dan cari masing-masing data yang ada kolom "antonim" (karena tidak semua kata memiliki antonim)
for word, attributes in thesaurus_data.items():
    if 'antonim' in attributes:
        words_with_only_antonim.append({
            'word': word,
            'antonim': ', '.join(attributes['antonim'])  #gabung antonim (kalau misal antonimnya ada lebih dari 1)
        })

#nama file csv untuk file antonim
csv_file_path = 'antonim_bahasa_indonesia.csv'

#write csv
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    #nama kolom "word" untuk kata dan "antonim" untuk antonimnya
    fieldnames = ['word', 'antonim']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    #header
    writer.writeheader()

    #isi
    for word_data in words_with_only_antonim:
        writer.writerow(word_data)

print(f"File berhasil dibuat '{csv_file_path}'.")
```
